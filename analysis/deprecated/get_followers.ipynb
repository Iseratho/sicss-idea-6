```python
import pandas as pd
```

```python
mps = pd.read_json("mps_flat.json", lines=True, dtype={"id": str})
mps
```

```python
mp_usernames = mps["username"].values
```

```python
import json
from tqdm import tqdm

with open("mp_followers.csv", "w") as file:
    for mp_user in tqdm(mp_usernames):
        val = !twarc2 following {mp_user} | twarc2 flatten
        for v in val:
            j = json.loads(v)
            if j["username"] in mp_usernames:
                file.write(f"{mp_user},{j['username']}\n")
```
